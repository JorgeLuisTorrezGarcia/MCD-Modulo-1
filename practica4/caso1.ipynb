# Reporte de Ventas 

Este reporte presenta un análisis de las ventas realizadas durante los tres primeros meses.

## 0. Importacion de Librerias

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

## 1. Recolección de Datos
Ingresamos los registros de las ventas diarias para el **Mes 1**, **Mes 2** y **Mes 3**.

```python
# DATOS DE VENTAS DIARIAS
# Ventas del mes 1
mes1 = [1200, 1500, 1800, 1100, 2000, 1750, 1600, 1900, 2100, 1700]

# Ventas del mes 2
mes2 = [1300, 1400, 1700, 1200, 2100, 1650, 1550, 1850, 2050, 1600]

# Ventas del mes 3
mes3 = [1400, 1600, 1750, 1300, 2200, 1800, 1700, 1950, 2150, 1750]
```

### Tabla Resumen de Ventas
Para facilitar su lectura, organizamos los datos diarios en una tabla, donde cada columna representa un mes y cada fila corresponde a un día de ventas.

```python
# Creamos un DataFrame para organizar la información
df = pd.DataFrame({
    "mes_1": mes1,
    "mes_2": mes2,
    "mes_3": mes3,
})

# Mostramos la tabla de ventas
display(df)
```

## 2. Promedio Diario por Mes
Calculamos cuánto se vendió *en promedio por día* en cada uno de los meses. Matemáticamente, sumamos las ventas de todos los días de un mes y las dividimos entre la cantidad de días registrados.

```python
# Calculamos el promedio diario de ventas por mes
mean_df = df.mean()

# Mostramos resultados de forma clara
print("Promedio diario de ventas por mes:\n")
print(mean_df)
```

## 3. Mes con Menor Ventas
Aquí sumamos todas las ventas diarias para obtener el **Total de ventas de cada mes**. Además, identificamos cuál de los meses tuvo el valor acumulado más bajo (mínimo).

```python
# Calculamos el total de ventas por mes
total_sales = df.sum()

# Identificamos el mes con menor ventas
min_month = total_sales.idxmin()
min_value = total_sales.min()
print(f"\nEl mes con menor ventas fue: {min_month}")
print(f"Total vendido: {min_value} Bs.")
```

## 4. Promedio y Mediana de las Ventas Totales
Calculamos dos indicadores estadísticos clave sobre las ventas totales mensuales calculadas en el paso anterior:
*   **Promedio mensual ($\\bar{x}$):** La suma de las ventas totales de los 3 meses dividida entre 3. Nos dice de manera general cuánto vendemos por mes.
*   **Mediana:** Si ordenamos el total de ventas mensual de menor a mayor, es el valor que queda exactamente en el medio.

```python
# Calculamos promedio y mediana de las ventas totales
mean_sales = total_sales.mean()
median_sales = total_sales.median()

print("Estadísticas de ventas totales:\n")
print(f"Promedio mensual de ventas: {mean_sales:.2f} Bs.")
print(f"Mediana de ventas mensuales: {median_sales:.2f} Bs.")
```

## 5. Crecimiento Trimestral
Para analizar el comportamiento de las ventas durante el trimestre se construye un gráfico de tendencia.
Este gráfico permite visualizar:

* La evolución de las ventas
* Posibles caídas o incrementos entre meses
* La tendencia general del desempeño comercial

```python
# Configuramos el tamaño del gráfico
plt.figure(figsize=(8,5))

# Gráfico de puntos
sn.scatterplot(
    x=total_sales.index,
    y=total_sales.values,
)

# Línea de tendencia que une los puntos
plt.plot(total_sales.index, total_sales.values)

# Títulos y etiquetas
plt.title("Tendencia de Ventas Totales por Mes")
plt.xlabel("Mes")
plt.ylabel("Ventas Totales (Bs)")

# Mostramos el gráfico
plt.show()
```
