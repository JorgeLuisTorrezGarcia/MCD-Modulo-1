<h2>importar paquetes</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2>Leer fuente de datos(archivo)</h2>

```python
df = pd.read_csv("ventas.csv")
df.head()
```

<h2>exploración inicial de datos</h2>

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

<h2>Limpieza basica de datos</h2>

<h3>convertir fecha a tipo datetime</h3>

```python
df["fecha"] = pd.to_datetime(df["fecha"])
```

<h3>crear una nueva columna: total de venta</h3>

```python
df["total_venta"] = df["precio"] * df["cantidad"]
df.head()
```

<h2>Analisis simple</h2>

<h3>ventas totales por categoria</h3>

```python
ventas_por_categoria = df.groupby("categoria")["total_venta"].sum()
ventas_por_categoria
```

<h2>Visualizacion de datos</h2>

```python
# grafico de barras
ventas_por_categoria.plot(kind="bar", figsize=(8,5))
plt.title("Ventas Totales por Categoría")
plt.xlabel("Categoría")
plt.ylabel("Total de Ventas")
plt.show()
```

### Conclusiones

- La categoría con mayores ventas es **Tecnología**.
- El análisis permite identificar qué tipo de productos generan mayor ingreso.
- Este tipo de análisis puede servir para toma de decisiones comerciales.
